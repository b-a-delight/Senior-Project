```python

def coefComp(area, df):
    k = 11
    df_Fun = df.copy()
    df_Fun = df_Fun[df_Fun['Area'] == area]
    df_Fun = trim(df_Fun)
    corrmat = df_Fun.corr()
    cols = corrmat.nlargest(k, 'TAV_Ch')['TAV_Ch']
    temp = pd.DataFrame(cols)

    return temp
```

```python
def trim(df):
    dfQ1 = df['TAV'].quantile(0.25)
    dfQ3 = df['TAV'].quantile(0.75)
    df_iqr = dfQ3 - dfQ1
    iqrMax =dfQ3 + (1.5*df_iqr)
    iqrMin = dfQ1 - (1.5*df_iqr)
    df_trim = df[(df['TAV']<iqrMax) & (df['TAV'] > iqrMin)]
    df_trim = df.drop(columns=['TAV'])
    
    return df_trim
```

```python
def newCC(df):
    k = 11
    df_Fun = df.copy()
    df_Fun = trim(df_Fun)
    corrmat = df_Fun.corr()
    cols = corrmat.nlargest(k, 'TAV_Ch')['TAV_Ch']
    temp = pd.DataFrame(cols)

    return temp
```

```python
def CCmerge(df,dfM, list):
    dfM =dfM
    for item in list:
        df = df[df['Area'] == item]
        df = trim(df)
        dfTemp = newCC(df)
        dfTemp = dfTemp.reset_index
        dfM = pd.merge(dfM, dfTemp, on = 'index', how = 'left')
        dfM = dfM.rename(columns={'TAV_Ch : item'})
        
    return dfM

```

## Experiment with removing outliers

```python
def ExCoefComp(area, df):
    k = 11
    df = df[df['Area'] == area]
    # remove outliers that are >3 std
    df = df[((df.TAV_Ch - df.TAV_Ch.mean()) / df.TAV_Ch.std()).abs() < 3]
    
    corrmat = df.corr()
    cols = corrmat.nlargest(k, 'TAV_Ch')['TAV_Ch']
    temp = pd.DataFrame(cols)

    return temp
```

```python

def NcoefComp(neigh, df):
    k = 11
    df_Fun = df.copy()
    df_Fun = df_Fun[df_Fun['Neighborhood'] == neigh]
    corrmat = df_Fun.corr()
    cols = corrmat.nlargest(k, 'TAV_Ch')['TAV_Ch']
    temp = pd.DataFrame(cols)

    return temp
```
